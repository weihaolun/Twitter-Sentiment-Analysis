```python
import pandas as pd
```

```python
data1 = pd.read_csv("resources/twitter-2013dev-A.txt", error_bad_lines=False, header=None)
data2 = pd.read_csv("resources/twitter-2013test-A.txt", error_bad_lines=False, header=None)
data3 = pd.read_csv("resources/twitter-2013train-A.txt", error_bad_lines=False, header=None)
data4 = pd.read_csv("resources/twitter-2014test-A.txt", error_bad_lines=False, header=None)
data5 = pd.read_csv("resources/twitter-2015test-A.txt", error_bad_lines=False, header=None)
data6 = pd.read_csv("resources/twitter-2015train-A.txt", error_bad_lines=False, header=None)
data7 = pd.read_csv("resources/twitter-2016dev-A.txt", error_bad_lines=False, header=None)
data8 = pd.read_csv("resources/twitter-2016devtest-A.txt", error_bad_lines=False, header=None)
data9 = pd.read_csv("resources/twitter-2016test-A.txt", error_bad_lines=False, header=None)
data10 = pd.read_csv("resources/twitter-2016train-A.txt", error_bad_lines=False, header=None)
```

```python
frames = [data1, data2, data3, data4, data5, data6, data7, data8, data9, data10]
initial_data_raw = pd.concat(frames)
```
