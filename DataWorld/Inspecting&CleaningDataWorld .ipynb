```python
import pandas as pd 
import numpy as np 
# Read each file as data frames.
dw_tweets_df = pd.read_csv("./Resources/text_emotion.csv", index_col=[0])
```

```python
dw_tweets_df.drop('author', axis=1, inplace=True)
```

```python
dw_tweets_df
```

```python
dw_tweets_df['sentiment'].unique()
```

```python
target = {
    'sadness': 0,
    'worry': 0,
    'hare': 0,
    'boredom': 0, 
    'anger': 0,
    'enthusiasm': 1,
    'love': 1,
    'fun': 1,
    'happiness': 1,
    'relief': 1
}
```

```python
dw_tweets_df["target"] = dw_tweets_df["sentiment"].map(target)
```

```python
dw_tweets_df
```

```python
dw_tweets_df.drop('sentiment', axis=1, inplace=True)
```

```python
dw_tweets_df
```

```python
dw_tweets_df.dropna()
```

```python
dw_pos_df = dw_tweets_df[dw_tweets_df['target'] > 0] 
```

```python
dw_pos_df
```

```python
dw_neg_df = dw_tweets_df[dw_tweets_df['target'] < 1] 
```
