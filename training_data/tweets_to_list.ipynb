```python
import pandas as pd
```

```python
# Read each file as data frames.
neutral_df = pd.read_csv("resources/filtered_data/neutral.csv", index_col=[0])
positive_df = pd.read_csv("resources/filtered_data/positive.csv", index_col=[0])
negative_df = pd.read_csv("resources/filtered_data/negative.csv", index_col=[0])
```

```python
# Extract "tweets" columns to lists.
neutral_list = neutral_df["tweets"].tolist()
positive_list = positive_df["tweets"].tolist()
negative_list = negative_df["tweets"].tolist()
```

```python
# Distribution of each type of attitudes.
print(f'There are total {len(neutral_list)} neuatral tweets, {len(positive_list)} positive tweets, and {len(negative_list)} negative tweets.')
```

```python
# Store lists to txt files as backup.
with open("resources/filtered_data/neutral_list.txt", "w") as output:
    output.write(str(neutral_list))

with open("resources/filtered_data/positive_list.txt", "w") as output:
    output.write(str(positive_list))

with open("resources/filtered_data/negative_list.txt", "w") as output:
    output.write(str(negative_list))
```
